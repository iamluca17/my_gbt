```python
import xgboost as xgb


# Create DMatrix for XGBoost (this is an optimized data structure used by XGBoost)
dtrain = xgb.DMatrix(redwine_train_f, label=redwine_train_o)
dtest = xgb.DMatrix(redwine_test_f, label=redwine_test_o)

# Set parameters for XGBoost
params = {
    'objective': 'multi:softmax',   # Multi-class classification
    'num_class': len(np.unique(redwine_train_f)), # Number of classes
    'eval_metric': 'merror',        # Classification error rate
    'learning_rate': 0.2,           # Same learning rate as your GBT
    'max_depth': 6,                 # Same depth for trees as your GBT (adjust as necessary)
    'colsample_bytree': 0.8,        # Feature subsampling (optional, tweak as needed)
    'subsample': 0.8,               # Row subsampling (optional, tweak as needed)
    'n_estimators': 50              # Same number of estimators as your GBT
}
```

```python

# Train the XGBoost model
num_round = 50  # Number of boosting rounds (same as number of estimators in GBT)
bst = xgb.train(params, dtrain, num_round)
```

```python

# Make predictions on the test set
preds = bst.predict(dtest)
```

```python
accuracy = accuracy_score(redwine_test_o, preds)
print(f"XGBoost Accuracy: {accuracy:.4f}")
```
